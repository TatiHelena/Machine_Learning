```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import mglearn
from cycler import cycler
```

```python
df = pd.read_csv('cycle-counts-2022.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Transform dtype object to datetime dtype. 
df['Time'] = pd.to_datetime(df['Time'])
df.info()
```

```python
df['Cycle_year'] = df['Time'].dt.year
df.sample(5)
```

```python
## To extract month from the date you use dt.month method and create new column with months

df['Cycle_Month_No'] = df['Time'].dt.month
df.sample(5)
```

```python
df['Cycle_Month'] = df['Time'].dt.month_name()
df.sample(5)
```

```python
## To extract day from the date you use dt.day method and create new column with day
df['Cycle_Day'] = df['Time'].dt.day
df.sample(5)
```

```python
## To extract day of the week from the date you use dt.day_name method and create new column with day of the week
df['Cycle_Dayofweek'] = df['Time'].dt.day_name()
df.sample(5)
```

```python
df['Cycle_Dayofweek'].replace('Monday', 1, inplace=True)
df['Cycle_Dayofweek'].replace('Tuesday', 2, inplace=True)
df['Cycle_Dayofweek'].replace('Wednesday',3, inplace=True)
df['Cycle_Dayofweek'].replace('Thursday', 4, inplace=True)
df['Cycle_Dayofweek'].replace('Friday', 5, inplace=True)
df['Cycle_Dayofweek'].replace('Saturday', 6, inplace=True)
df['Cycle_Dayofweek'].replace('Sunday', 7, inplace=True)
```

```python
df.sample(10)
```

```python
## To extrat if cyclists are cycling on Saturday or Sunday 
df['Cycle_On_Weekend'] = np.where(df['Cycle_Dayofweek'].isin(['Sunday','Saturday']), 1,0 )
df.sample(10)
```

```python
df['Hour'] = df['Time'].dt.time
df.head()
```

```python
df.info()
```

```python
# Features of my dataset
X = new.iloc[:,:31]

# Store the target variable as lable into an array y
y = new.iloc[:,31]

# Display number of rows and columns
X.shape, y.shape
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) 

X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
# 0.2 means 20% for testing and random_state simply sets a seed to the random generator, so that your train-test splits
# are always deterministic. If we don't set a seed, it is different each time.
```

```python
from sklearn.preprocessing import StandardScaler   #We scale the data in order to calculate the correct distance between datapoints. 

# Create an object using Standard scalscaler
scaler = StandardScaler()

# Train the StandardScaler model
scaler.fit(X_train)

# transform data
X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)
X_scaled = scaler.transform(X)

# print dataset after scaling
print(X_train_scaled, '\t', X_test_scaled, '\t', X_scaled)
```

```python
from sklearn.decomposition import PCA

# keep the first two principal components of the data
pca = PCA(n_components = 3)

# transform data into the first two principal components
X_pca = pca.fit_transform(X_scaled)

print("Original shape: {}".format(str(X_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))
```

```python
print(pca.explained_variance_ratio_)
```
