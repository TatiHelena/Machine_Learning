## Implementing LDA 

LDA is a supervised ML algorithm used specifically for classification tasks in Machine Learning. 
Differently from PCA, LDA find the decision boundaries that best separates the classes in a dataset. 

LDA criteria:
1.	Maximize the distance between means of different classes
2.	Minimize the variation within each class

It will generate a new axis in which increase the distance between the data points of different classes. Then finding the directions in the feature space that best separate the different classes of the data. (Raman, 2019).


```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
easy_lda = lda.fit_transform(X, y)
```

```python
easy_lda
```

```python
lda.explained_variance_ratio_
```

```python
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.scatter(
    easy_lda[:,0],
    easy_lda[:,1],
    c=y,
    cmap='rainbow',
    alpha=0.7,
    edgecolors='b')
```

## Implementing PCA

Because the curse of dimensionality of the dataset (higher number of features), the PCA (Principal Component Analysis) tool has been applied in order to reduce the dataset dimensionality and retains the most important information of the dataset.

After implementing PCA a new dataset has been created with new features with (2 principal components) in which contains the most variance ratio (best fit) of the dataset, the variation rate has been of 0.91%.  

It has been shown below that the first component C1 has 0.7874%% of variance of the NHANES dataset while the second component C2 has only 0.1172%. Although it has been picked 2 components for the new dataset after PCA, I believe that work with one component would be sufficient for the analysis as it retains almost 80% of the dataset info. 


```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

# Aplying Machine Learning
Splitting the data

In Supervised Machine Learning the goal is to build a model with the training data, that is able to do the predctions of new, unseen data that has the same characteristic of the training data that has been used before. It is important that the model can make good generalization from the training set to the test set, which means that the model can make accurate predictions of new data. The evaluation of the test set will tell if a model has been performed well or not.

In order to set up our training and test. The NHANES dataset which contains  observations, will be trained on ML model using the method Holdout, in which the dataset has been divided in two random parts: a training and testing set. The split will be in the ratio of 70% for training and the remaining 30% of the data will be used for testing.

There is also the terms called Overfitting which means that the ML model has learned/memorized the training data very well but fails to perform on the testing data, also Underfitting when there isn't enough variety of data and for that reason the model perform poorly on the training set.



# Supervised, KNN model 
KNN is a supervised ML algorithm that can be used for classification and regression. KNN is also known as a "lazy leaner" algorithm, the reason it is because KNN does not learn a particular function of the training data, but instead it memorizes. 
The model classify a data point based in the nearest group of data points it belongs to according to the number of ‘Ks’. The distance between the data points  is calculated using Euclidean distance. 
The chosen number of Ks will impact on the classification of datapoints. A small ‘K’ value can be noisy and more likely to outliers, while selecting a large value can overlook categories with fewer datapoints.                                     (Shafi, 2023)                                                   



```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state=1)
```

```python
# Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

# Create KNN Classifier
kNN = KNeighborsClassifier(n_neighbors = 2)

# Train the model using the training sets
kNN.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = kNN.predict(X_test)
```

```python
print("Training set score: {:.2f}".format(kNN.score(X_train, y_train)))
print("Test set score: {:.2f}".format(kNN.score(X_test, y_test)))
```

# Unsupervised ML, Clustering - K-Means 

Clustering means to group similar data points into distinct groups or clusters. The clustering algorithm can be used for an analysis or as a feature engineering tool in order to better supervised ML models. 

What it basically does is, suppose there are groups of data points, then for each one of these groups there is a center point identifier, which will be used to refer and describe those groups. 

Examples of clustering can be: clustering similar products or clustering customers for different market segmentation. 

In order to find the similarity of the data points to group them into the same cluster, the algorithm can use distance measures for instance, Euclidean, Minkowski or Manhattan. 

For NHANES analysis it will use K-Means clustering model which is a unsupervised ML model, means that the datapoints are not label.


## Finding the number of clusters:

There are different methods to find the number of clusters in a dataset, but not every clustering algorithm needs a specific number of clusters. In the case of KMeans, that is necessary. It has been applied two methods to find the best number of clusters in the NHANES dataset. 
(Avinash Navlani, Fandango and Idris, 2021)


## Elbow Method 

It basically choose the number of clusters focusing on the percentage of variance for each different cluster. Using Within-Cluster Sum of Squares (WCSS), also known as inertia. 

we can find the ideal number of cluster using the Elbow Method. The number os cluster will be chosen in the point where it starts to flatten or improve performance. For this dataset, k=3 is probably the best option. 
(Avinash Navlani, Fandango and Idris, 2021)


```python

from sklearn.cluster import KMeans 
wcss_list = []


# Run a loop for different value of number of cluster
for i in range(1, 9):
    
 # Create and fit the KMeans model
    kmeans_model = KMeans(n_clusters = i, random_state = 123)
    kmeans_model.fit(X_pca)

 # Add the WCSS or inertia of the clusters to the score_list
    wcss_list.append(kmeans_model.inertia_)

# Plot the inertia(WCSS) and number of clusters
plt.plot(range(1, 9), wcss_list, marker='*')

# set title of the plot
plt.title('Selecting Optimum Number of Clusters using Elbow Method')

# Set x-axis label
plt.xlabel('Number of Clusters K')

# Set y-axis label
plt.ylabel('Within-Cluster Sum of the Squares(Inertia)')

# Display plot
plt.show()
```

## Sillouette 

The Sillouette method will indicates how well a data point is grouped together in their own clusters and separated from the others. 
The method scores range from -1 to +1. if a value is positive that means the cluster is well separated, in the case of a negative value means a datapoint has been assigned incorrectly. (Avinash Navlani, Fandango and Idris, 2021)

Evaluating the number of clusters using Sillouette, we see that k=2 would be probably the best choice. 

```python
from sklearn.metrics import silhouette_score
score_list = []

# Run a loop for different value of number of cluster
for i in range(2, 8):
    
 # Create and fit the KMeans model
     kmeans_model = KMeans(n_clusters = i, random_state = 123)
     kmeans_model.fit(X_pca)
 # Make predictions
     pred=kmeans_model.predict(X_pca)
 # Calculate the Silhouette Score
     score = silhouette_score (X_pca, pred, metric='euclidean')
 # Add the Silhouette score of the clusters to the score_list
     score_list.append(score)
    
# Plot the Silhouette score and number of cluster
plt.bar(range(2, 8), score_list)

# Set title of the plot
plt.title('Silhouette Score Plot')

# Set x-axis label
plt.xlabel('Number of Clusters K')

# Set y-axis label
plt.ylabel('Silhouette Scores')

# Display plot
plt.show()

```

```python


#Define number of clusters
num_clusters = 3

# Create and fit the KMeans model
km = KMeans(n_clusters=num_clusters)
km.fit(X_pca)

# Predict the target variable
pred=km.predict(X_pca)

# Plot the Clusters
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=pred, marker="o", cmap="bwr_r")

# Set title of the plot
plt.title('K-Means Clustering')

#Set x-axis label
plt.xlabel('X-Axis Values')

# Set y-axis label
plt.ylabel('Y-Axis Values')

# Display the plot
plt.show()
```

# Model Evaluation

In order to evaluate the KMeans model it has been used different methods which are categorized into internal and external performance evaluation. 
   
### Internal Performance evaluation. 
It is based on the data features and does not include the target label data. It also will give a better score for clustering methods in which clusters are well separated. But it does not mean that a higher score has a better performance of the algorithm. 
    
- The Davies-Bouldin index (BDI) - Ratio of intra-cluster distance to inter-cluster distance. Getting a lower score of DBI means a better model performance. 

- The silhouette coefficient - This method also used to identify the number of clusters will also the similarity od a data point within its own cluster and other nearest groups. A higher score of the Sillouette means a better performance of the algorithm 

### External Performance evaluation
It is based used the target feature of the dataset that has not been used in the KMeans process. It has a similar process of supervised ML evaluation, that means that we can use confusion matrix to evaluate model results. these measures will evaluate the quality of created clusters. 

- The Rand score - it will evaluate the similarity between a cluster and the standard. 

- The Jaccard score - it will evaluate the similiarity of two datasets. It scores range from 0 to 1, which 0 means the dataset has no similar elements, and 1 means the datasets are alike. A lower score is better because it means the representation of different clusters. 

- F-Measure or F1-score - it will evaluate precision and model's robustness. It is balanced mean of precision and recall values, also it used B-value to equalize the representation of false negatives. 

- The Fowlkes-Mallows score - it is a geometric mean of precision and recall. A lower score represents distinct clusters. 
(Avinash Navlani, Fandango and Idris, 2021)



```python
# partition data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=1)


# Import metrics module for performance evaluation
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import fowlkes_mallows_score



# Specify the number of clusters
num_clusters = 3

# Create and fit the KMeans model
km = KMeans(n_clusters=num_clusters)
km.fit(X_train)

# Predict the target variable
predictions = km.predict(X_test)

# Calculate internal performance evaluation measures
print("Davies-Bouldin Index:", davies_bouldin_score(X_test,
predictions))
print("Silhouette Coefficient:", silhouette_score(X_test,
predictions))

# Calculate External performance evaluation measures
print("Adjusted Rand Score:", adjusted_rand_score(y_test,
predictions))
print("Jaccard Score:", jaccard_score (y_test, predictions, average='micro'))
print("F-Measure(F1-Score):", f1_score(y_test, predictions, average='micro'))
print("Fowlkes Mallows Score:", fowlkes_mallows_score(y_test,
predictions))
```

### Conclusion 

Based on the Elbow method, the ideal number of clusters would be 3, while the Sillouete model shows that the number of 2 clusters would be the best choice. 

In this case, KMeans model has been evaluated with 2 and 3 clusters using different internal and external performance measures. According to internal evaluation measures DBI and Sillouette Coefficient the KMeans model with 2 clusters has a better performance than the one with 3 clusters. On the other hand external performance measures like Adjusted Rand Score and Adjusted Rand Score has considered KMeans with 3 clusters having better results.

It has been observed in the figure 22 and 23 above that KMeans model has not been able to separate the clusters properly and it has not drawn a boundary between them, which indicates that the algorithm has failed to perform clustering task.
